```python
def quicksort(tab):
    if len(tab) <= 1:
        return tab
    lim=tab[0]
    less=[x for x in tab if x<lim]
    equal=[x for x in tab if x==lim]
    plus=[x for x in tab if x>lim]
    return quicksort(less)+equal+quicksort(plus)

n=int(input())
A=list(map(int,input().split()))
result=quicksort(A)
print(' '.join(map(str,result)))
```
