```python
def heap(arr, n, i):
    largest=i
    l=2*i+1
    r=2*i+2
    if l<n and arr[i]<arr[l]:
        largest=l
    if r<n and arr[largest]<arr[r]:
        largest=r
    if largest!=i:
        arr[i],arr[largest]=arr[largest],arr[i]
        heap(arr,n,largest)

def maxh(arr):
    n=len(arr)
    for i in range(n//2-1,-1,-1):
        heap(arr,n,i)
    return arr

n=int(input())
A=list(map(int,input().split()))
result=maxh(A)
print(' '.join(map(str,result)))
```
