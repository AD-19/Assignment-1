```python
n=int(input())
A=list(map(int,input().split()))
m=int(input())
B=list(map(int,input().split()))

i=j=0
C=[]

while i<len(A) and j<len(B):
    if A[i]<B[j]:
        C.append(A[i])
        i+=1
    else:
        C.append(B[j])
        j+=1

C.extend(A[i:])
C.extend(B[j:])

print(' '.join(map(str,C)))
```
